While the main purpose of the study was analyze what the determinants of development institutions are, our European colleague was also interested at looking at the individuals who worked *in* these institutions to see if they differed from the general Portuguese population with regards to their values. To measure this we used [Schwartz Basic Human Values Theory](https://scholarworks.gvsu.edu/cgi/viewcontent.cgi?article=1116&context=orpc) and the subset of [21 questions](https://www.europeansocialsurvey.org/docs/methodology/core_ess_questionnaire/ESS_core_questionnaire_human_values.pdf) from Schwartz's original 56 that are included in the European Social Survey. 

Before including these measures and the scales derived from them in our final publication, I was tasked with checking that the data reflected what the theory proposed. This validation was one of the most enjoyable parts of the project for me since I got to learn a new theory **and** a new method, non-metric multidimensional scaling. Unfortunately, the work I did was never published, to my knowledge, so I will take this opportunity to share it with you.

## Data and Analysis 

The study focused of six institutions, four of which we have survey data for.  The institutions are: 
* The Energy Corporation (Energias de Portugal - EDP) 
* The Food and Economic Security Authority (Autoridade de Segurança Alimentar e Económica – ASAE) 
* The Postal Service (Correios de Portugal – CTT) 
* The Public Health System; Santa Maria Hospital (Serviço Nacional de Saúde; Hospital de Santa Maria - HSM))

We want to make sure that the data from each institution we are studying reflect the theorized spatial relationship between the values. That is, the variables we have that measure the values should reflect the relationship in the image above, e.g. measures of Universalism should be near the measures of Self-Direction and Beneveloence. 

The data can be found [here](https://cmd.princeton.edu/publications/data-archives/viqd). For this Notebook, I use the data I have from the project, which was used to create the data found in the link. The data should be the same except that my data is already converted into numeric values and the data in the link is categorical.   

I follow what Schartz did and use [non-metric multidimensional scaling](https://en.wikipedia.org/wiki/Multidimensional_scaling#Non-metric_multidimensional_scaling) to plot how the different variables relate to each other in a two-dimensional space. The basic idea is to find how distant or far apart each variable is from the others and plot it on a two-dimensional plane. To calculate how distant each variable is from the others, we find the [euclidean distance](https://en.wikipedia.org/wiki/Euclidean_distance) of each variable. Our points are the variables and coordinates used to find the distance are the values of the variables for each observation. We then use non-metric multidimensional scaling to visualize the points. The Python code is presented below. This was translated from original code in Stata.

```python
#Load libraries. 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
#Load the data. 
viqd = pd.read_stata('C:\\Users\\jeann\\Desktop\\viqd_all_inst.dta') 
```

The variables are saved as v31 to v51 in the data set. Let's select only those as well as the variable for institution. Next, let's drop any observation with missing values on any of the questions. 

```python
#Get list of variables we need. 
var = ["v" + str(num) for num in np.arange(31, 52, 1)]  
var.append('inst')

#Get only variables we need. 
viqd = viqd[var] 

#Drop missing variables. 
viqd.dropna(how = 'any', inplace = True)
```

For illustrative purposes, I will show what I did for one institution, the Food and Economic Security Authority (ASAE), in depth. I will then write a function that does everything to make plotting easier. 

```python
#Create separate data sets. 
asae = viqd.loc[viqd['inst'] == 'asae'] 
asae.head()
```

To make interpretation of the plot easier, I rename the variables. 

```python
asae = asae.iloc[:,:-1]
cols = ['create','rich','equal','ablty','safe','diff',
        'frule','udrst','modst','gdtim','free','hlppl',
        'suces','strgv','advnt','bhv','rspct','lyl',
        'env','trad','fun']  

asae.columns = cols 

asae.head()
```

I load the necesarry packages, one to compute euclidean distance and one for multidimesional scaling to find the points in a two-dimensional space. 

```python
from sklearn.manifold import MDS  
from sklearn.metrics.pairwise import euclidean_distances 

#Find Euclidean distance.  
d = euclidean_distances(asae[cols].T) 
```

In this next part I find the positions of the points using non-metric multidimensional scaling (nmds), and specify some of the arguments for the nmds function. 

```python
#Use KJR seed.
mds = MDS(n_components=2, max_iter=3000, eps=1e-9, random_state=102494,
                   dissimilarity="precomputed", n_jobs=1)

pos = mds.fit(d).embedding_

nmds = MDS(n_components = 2, metric = False, random_state = 102494, dissimilarity = 'precomputed').fit_transform(d, init = pos)
```

This code plots the results. 

```python
plt.figure(figsize = (12,9))
plt.scatter(nmds[:,0], nmds[:,1], color = 'white') 
for i in np.arange(0, 21, 1): 
    plt.text(nmds[i,0], nmds[i,1], cols[i], fontsize = 15)

plt.xlabel('Dimension 1', size = 15) 
plt.ylabel('Dimension 2', size = 15) 
plt.title('Non-Metric Multidimensional Scaling Results for Schwartz Value Scale Variables: ASAE', size = 15) 
```

The image above plots the measures in a two-dimensional space. With this, we can make a judgment about the validity of the theorized model. As someone who has worked with this data, I can say that this plot more-or-less shows what we would expect to see. However, you are likely not familiar with this data, so that is not as clear to you. To make the visualization easier to understand, I replace the variable names with the names of the values they are measuring. I also add color as a dimension with Universalism as the lightest blue and Power as the darkest blue. Thus, we should see darker colors close to each other and away from lighter colors.  

The code below creates dictionaries that map variable names to value names and value names to colors.  The code next bit of code  contains a function that puts everything together.  

```python
a = ['Universalism', 'Benevolence', 'Tradition', 'Conformity', 'Security',
     'Power', 'Achievement', 'Hedonism', 'Stimulation', 'Self-Direction']

b = ['PowderBlue', 'SkyBlue', 'DeepSkyBlue','CornflowerBlue','RoyalBlue',
     'DarkBlue','RoyalBlue','CornflowerBlue','DeepSkyBlue','SkyBlue'] 

colors = dict(zip(a,b)) 

cols = ['create','rich','equal','ablty','safe','diff',
        'frule','udrst','modst','gdtim','free','hlppl',
        'suces','strgv','advnt','bhv','rspct','lyl',
        'env','trad','fun']  

vals = ['Self-Direction','Power','Universalism','Achievement','Security','Stimulation', 
 'Conformity','Universalism','Tradition','Hedonism','Self-Direction','Benevolence',
 'Achievement','Security','Stimulation','Conformity','Power','Benevolence',
 'Universalism','Tradition','Hedonism'] 

var_to_vals = dict(zip(cols,vals))
```

```python
def plot_nmds(df, inst, inst_name): 
    
    #Subset data. 
    df = df.loc[df['inst'] == inst] 
    df = df.iloc[:,:-1]
     
    #Drop missing values. 
    df.dropna(how = 'any', inplace = True) 
    
    #Load packages.
    from sklearn.manifold import MDS  
    from sklearn.metrics.pairwise import euclidean_distances 
    
    #Find dissimilarity, Euclidean distance. 
    d = euclidean_distances(df.T)
    
    mds = MDS(n_components=2, max_iter=3000, eps=1e-9, random_state=102494,
                   dissimilarity="precomputed", n_jobs=1) 
    pos = mds.fit(d).embedding_
    nmds = MDS(n_components = 2, metric = False, random_state = 102494, 
               dissimilarity = 'precomputed').fit_transform(d, init = pos)
    
    #PLot nmds. 
    plt.figure(figsize = (12,9))
    plt.scatter(nmds[:,0], nmds[:,1], color = 'white') 
    
    #Plot text.
    for i in np.arange(0, 21, 1):  
        vals_name = var_to_vals.get(cols[i] ) 
        color = colors.get(vals_name)
        plt.text(nmds[i,0], nmds[i,1], vals_name, fontsize = 15, color = color)
    
    #Add labels. 
    ttl = 'Non-Metric Multidimensional Plot for ' + inst_name
    plt.xlabel('Dimension 1', size = 15) 
    plt.ylabel('Dimension 2', size = 15) 
    plt.title(ttl, size = 15)
    
    
```

```python
plot_nmds(viqd, 'asae','The Food and Economic Authority (ASAE)')
```

```python
plot_nmds(viqd, 'ctt', 'The Postal System (CTT)')
```

```python
plot_nmds(viqd, 'edp', 'The Energy Corporation (EDP)')
```

```python
plot_nmds(viqd, 'hsm', 'Santa Maria Hospital (HSM)')
```
